```python
import matplotlib.pyplot as plt
```

```python
n_clusters = [1, 5, 10, 20, 30, 50, 100, 150] # for 40km and 25km
# n_clusters = [1, 5, 10, 30, 50, 75, 100, 150, 200] # for 10km
err_40km = [1.321428, 0.735714, 0.603571, 0.65714, 0.632143, 0.675, 0.671429, 0.63214]
err_25km = [1.321053, 0.773684, 0.639474, 0.634211, 0.652632, 0.644737, 0.702632, 0.639474]
err_10km = [1.334286, 0.708571, 0.657143, 0.714286, 0.64, 0.628571, 0.622857, 0.585714, 0.582857]
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-v0_8-deep")  # Use a clean style

resolution = '40km'

err = err_40km
if resolution == '25km':
    err = err_25km
if resolution == '10km':
    err = err_10km

plt.figure(figsize=(6, 4), dpi=300)  # Increase figure size & resolution
plt.plot(n_clusters, err, marker='o', linestyle='-', markersize=5, label='Mean Absolute Error')

plt.xlabel('Number of Clusters', fontsize=9, fontweight='bold')
plt.ylabel('Mean Absolute Error Over Clusters', fontsize=9, fontweight='bold')
plt.title(f'MAE vs. Number of Clusters\nResolution: {resolution} Pixels', fontsize=11, fontweight='bold')

# Add major and minor ticks
ax = plt.gca()
ax.xaxis.set_major_locator(plt.MultipleLocator(20))  # Major ticks every 20
ax.xaxis.set_minor_locator(plt.MultipleLocator(10))  # Minor ticks every 10

# Customize tick labels
ax.tick_params(axis='x', which='major', labelsize=10, width=1.2, length=6)  # Larger major ticks
ax.tick_params(axis='x', which='minor', labelsize=8, width=0.8, length=4, color='gray')  # Smaller minor ticks

ax.tick_params(axis='y', which='major', labelsize=10)  # Ensure y-axis is readable

# Add grid lines
ax.grid(True, which='major', linestyle='--', alpha=0.6)  # Major grid (dashed)
ax.grid(True, which='minor', linestyle=':', alpha=0.4)  # Minor grid (dotted)

plt.legend()

plt.show()
```

```python
# Cluster values for different resolutions
n_clusters_40_25km = [1, 5, 10, 20, 30, 50, 100, 150]  # For 40km and 25km
n_clusters_10km = [1, 5, 10, 30, 50, 75, 100, 150, 200]  # For 10km

# Mean absolute errors (MAE) for each resolution
err_40km = [1.321428, 0.735714, 0.603571, 0.65714, 0.632143, 0.675, 0.671429, 0.63214]
err_25km = [1.321053, 0.773684, 0.639474, 0.634211, 0.652632, 0.644737, 0.702632, 0.639474]
err_10km = [1.334286, 0.708571, 0.657143, 0.714286, 0.64, 0.628571, 0.622857, 0.585714, 0.582857]

plt.style.use("seaborn-v0_8-deep")  # Use a clean style

# Create figure
plt.figure(figsize=(6, 4), dpi=300)

# Plot all three curves with different markers and styles
plt.plot(n_clusters_10km, err_10km, marker='o', linestyle='-', markersize=3, label='10km Resolution')
plt.plot(n_clusters_40_25km, err_40km, marker='^', linestyle='-.', markersize=3, label='40km Resolution')
plt.plot(n_clusters_40_25km, err_25km, marker='s', linestyle='--', markersize=3, label='25km Resolution')

# Labels and title
plt.xlabel('Number of Clusters', fontsize=9, fontweight='bold')
plt.ylabel('Mean Absolute Error Over Clusters', fontsize=9, fontweight='bold')
plt.title('MAE vs. Number of Clusters for Different Resolutions', fontsize=11, fontweight='bold')

# Grid and tick settings
ax = plt.gca()
ax.xaxis.set_major_locator(plt.MultipleLocator(20))  # Major ticks every 20
ax.xaxis.set_minor_locator(plt.MultipleLocator(10))  # Minor ticks every 10
ax.tick_params(axis='x', which='major', labelsize=10, width=1.2, length=6)  # Major ticks
ax.tick_params(axis='x', which='minor', labelsize=8, width=0.8, length=4, color='gray')  # Minor ticks
ax.tick_params(axis='y', which='major', labelsize=10)  # Ensure y-axis is readable

# Add grid lines
ax.grid(True, which='major', linestyle='--', alpha=0.6)  # Major grid (dashed)
ax.grid(True, which='minor', linestyle=':', alpha=0.4)  # Minor grid (dotted)

# Add legend
plt.legend(fontsize=9)

# Show plot
plt.show()
```
