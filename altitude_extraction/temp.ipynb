```python
# Set up figure with better dimensions for publication
plt.figure(figsize=(10, 6), dpi=150)

cutoff_idx = n_clusters.index(30) + 1

# Color palette suitable for publication
colors = ['#0072B2', '#D55E00', '#009E73', '#CC79A7', '#F0E442', '#56B4E9', '#E69F00', '#000000']
markers = ['o', 's', '^', 'D', 'v', '<', '>', 'p']
linestyles = ['-', '--', '-.', ':', '-', '--', '-.', ':']

# Create the plot with improved styling
resolutions = [150, 130, 100, 80, 60, 40, 25, 10]
all_data = [res150km, res130km, res100km, res80km, res60km, res40km, res25km, res10km]
all_data = [data[:cutoff_idx] for data in all_data]

for i, (res, data) in enumerate(zip(resolutions, all_data)):
    plt.plot(n_clusters[:cutoff_idx], data, color=colors[i], linestyle=linestyles[i], 
             marker=markers[i], markersize=6, linewidth=1.5, 
             markerfacecolor='white', markeredgewidth=1.5,
             label=f'{res} km')

# Set better x-axis ticks
major_ticks = [1, 5, 10, 15, 20, 25, 30]
minor_ticks = list(range(0, 36))
plt.xticks(major_ticks)
plt.gca().set_xticks(minor_ticks, minor=True)

# Set bounds
plt.xlim(0, 30.7)

# Set better y-axis ticks
plt.yticks(np.arange(0, 5.5, 0.5))
plt.gca().set_yticks(np.arange(0, 5.5, 0.1), minor=True)

# Add grid for readability
plt.grid(which='major', linestyle='-', linewidth=0.5, color='gray', alpha=0.7)
plt.grid(which='minor', linestyle=':', linewidth=0.5, color='gray', alpha=0.4)


# Improve labels and title
plt.xlabel('Number of Clusters', fontsize=12, fontweight='bold')
plt.ylabel('Point-Averaged Squared Error', fontsize=12, fontweight='bold')
plt.title('Squared Error vs Number of Clusters', fontsize=14, fontweight='bold')

# Better legend
legend = plt.legend(title='Resolution', fontsize=10,
                    frameon=True, fancybox=True, shadow=True,
                    loc='center left', bbox_to_anchor=(1, 0.5))
legend.get_title().set_fontweight('bold')

# Tight layout and adjust margins
plt.tight_layout()
plt.subplots_adjust(right=0.85)  # Make room for legend

plt.show()
```
